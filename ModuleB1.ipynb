17.1) Interpolation Problem Statement

```python
import numpy as np # importing the libraries
import matplotlib.pyplot as plt

x = [2, 4, 6]
y = [6, 4, 0]

x_n = 5
y_n = np.interp(x_n, x, y)
print(y_n) # priniting the new value of y using the interpolation formula
```

```python
plt.plot(x, y, '-k')
plt.title("Interpolation of x at the value 5")
plt.xlabel('x-axis') 
plt.ylabel('y-axis')
plt.show()
```

17.2) Linear Interpolation

```python
# imported the libraries
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
```

```python
# the points on both the x-axis and the y-axis
x = [1, 2, 4] 
y = [2, 6, 8]

a = interp1d(x, y)
y_h = a(3)
print(y_h)
```

```python
# declared the figure size and plotted the graph for linear interpolation
plt.figure(figsize = (10, 10))
plt.plot(x, y, '-ob')
plt.plot(3, y_h, 'ro')
plt.title("Linear interpolation of x at the value 5")
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

17.3) Cubic Spline Interpolation

```python
from scipy.interpolate import CubicSpline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [4, 6, 3]

b = CubicSpline(x, y, bc_type = 'natural')
x_n = np.linspace(0, 4, 100)
y_n = b(x_n)
print(y_n)
```

```python
plt.figure(figsize = (15, 15))
plt.plot(x_n, y_n, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```
