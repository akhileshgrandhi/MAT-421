17.4 Lagrange Polynomial Interpolation
In this part, we will discuss some examples of Lagrange Polynomial Interpolation and code that example in python.

Lagrange Polynomial Interpolation finds a single polynomial that goes through all the data points.

```python
# importing the libraries

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange
```

```python
x = [0, 2, 4]
y = [1, 6, 4]

x_n = np.arange(0, 5, 0.1)

f = lagrange(x,y)
```

```python
fig = plt.figure(figsize = (10, 10))
plt.plot(x_n, f(x_n), 'b', x, y, 'ro')
plt.title("Lagrange Polynomial Interpolation")
plt.grid()
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

Let's do other cases and different points

```python
x_two = [4, 5, 6]
y_two = [5, 9, 7]

x_n = np.arange(0, 10, 0.5)
f_two = lagrange(x_two, y_two)
```

```python
plt.plot(x_n, f_two(x_n), 'b', x_two, y_two, 'ro')
plt.title("Lagrange Polynomial Interpretation")
plt.grid()
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

17.5 Newton's Polynomial Interpolation In this part, we will discuss some examples of Newton's Polynomial Interpolation and code that example in python.

Newton's Polynomial Interpolation is an another way to fit exactly for the given number of data points.

```python
# importing all the libraries
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')

%matplotlib inline
```

```python
def divided_diff(x, y):

  n = len(y)
  coef = np.zeros([n, n])
  coef[:,0] = y

  for j in range (1,n):
    for i in range(n-j):
      coef[i][j] = \
      (coef [i+1][j-1] - coef[i][j-1]) / (x[i+j]-x[i])

  return coef

def newton_poly(coef, x_data, x):

  n = len(x_data) - 1
  p = coef[n]
  for k in range (1, n+1):
    p = coef[n-k] + (x - x_data[n-k]) * p
  return p
```

```python
x_three = np.array([-5, -3, 0, 3])
y_three = np.array([-4, 4, 1, 4])

a_s = divided_diff(x_three, y_three) [0, :]

x_new = np.arange(-6, 3.5, .1)
y_new = newton_poly(a_s, x_three, x_new)

fig = plt.figure(figsize = (10, 10))
plt.plot(x_three, y_three, 'bo')
plt.plot(x_new, y_new)
     
```
